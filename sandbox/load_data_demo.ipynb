```python
import os

def list_subdirectories(path: str) -> None:
    """
    Lists only the files and subdirectories in the given directory (non-recursive).

    Parameters:
        path (str): The directory path to inspect.

    Returns:
        None
    """
    try:
        with os.scandir(path) as entries:
            for entry in entries:
                if entry.is_file():
                    print(f"File: {entry.name}")
                elif entry.is_dir():
                    print(f"Subdirectory: {entry.name}")
    except FileNotFoundError:
        print("The specified path does not exist.")
    except PermissionError:
        print("Permission denied to access this path.")
```
