# Matrix Profile based Anomaly Detection for ECG Seizure Detection


```python
import numpy as np
import stumpy
import matplotlib.pyplot as plt
```

```python
def calculate_matrix_profile_for_sample(sample:np.array, subsequence_length:int):
    return stumpy.stump(sample, subsequence_length)
```

```python
def get_annomaly_index_from_matrix_profile(matrix_profile:np.ndarray) -> int:
    return np.argsort(matrix_profile[:, 0])[-1]
```

```python
def plot_matrix_profile(sample: np.ndarray, matrix_profile: np.ndarray) -> plt.Figure:
    fig, ax = plt.subplots(2, 1, figsize=(12, 4))

    ax[0].plot(sample)
    ax[0].set_title("time series")

    ax[1].plot(matrix_profile[:, 0])
    # ax[1].set_title(f"Found outlier at {annomaly_index()}")
    ax[1].set_title(f"Found outlier at {get_annomaly_index_from_matrix_profile(matrix_profile=matrix_profile)}")

    fig.text(0.01, 0.01, "Matrixprofile (small = similar, large = potential outlier)", fontsize=8, color="gray")
    fig.tight_layout()
    plt.show()
```

```python
timeseries = np.array([1, 1, 1, 1, 1, 9, 1,1,1,1,1,1,1], dtype=np.float64)
subsequence_length = 3
matrix_profile=calculate_matrix_profile_for_sample(sample=timeseries, subsequence_length=subsequence_length)
print(matrix_profile)
plot_matrix_profile(sample=timeseries, matrix_profile=matrix_profile)
```

```python
timeseries = np.array([1, 1, 3, 1, 2, 9, 1,5,1,6,1,7,1], dtype=np.float64)
subsequence_length = 3
matrix_profile=calculate_matrix_profile_for_sample(sample=timeseries, subsequence_length=subsequence_length)
print(matrix_profile)
plot_matrix_profile(sample=timeseries, matrix_profile=matrix_profile)
```
