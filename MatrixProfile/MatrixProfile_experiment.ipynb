# Matrix Profile based Anomaly Detection for ECG Seizure Detection


```python
import numpy as np
```

```python
# from matrixprofile import *
```

```python
import stumpy
```

```python
# ts = np.array([0, 1, 2, 3, 2, 1, 0], dtype=np.float64)
ts = np.array([1, 1, 1, 1, 1, 9, 1,1,1,1,1,1,1], dtype=np.float64)
m = 4 # subsequence length
mp = stumpy.stump(ts, m)
```

```python
mp
```

```python
discord_idx = np.argsort(mp[:, 0])[-1]
print(f"Discord (Anomalie) bei Index {discord_idx}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 4))

plt.subplot(2, 1, 1)
plt.plot(ts)
plt.title("Zeitreihe")

plt.subplot(2, 1, 2)
plt.plot(mp[:, 0])
plt.title("Matrixprofil (klein = Ähnlichkeit, groß = Ausreißer)")

plt.tight_layout()
plt.show()
```
